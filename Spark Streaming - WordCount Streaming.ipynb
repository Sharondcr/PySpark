```python
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
from pyspark.sql import SparkSession
from pyspark import SparkConf,SparkContext
```

### 1.Create SparkStreaming Context

```python
#intialized a new spark session
config = SparkConf().setAppName('SparkSession').setMaster('local[4]')
sc = SparkContext.getOrCreate(conf=config)
```

```python
#Intialize Steaming context with batch interval of 5 seconds

ssc = StreamingContext(sc, 5)
```

```python
#Create a DStream that will connect to URL with localhost as IP & Port Number as 9990

lines = ssc.socketTextStream("localhost",8001)
words = lines.flatMap(lambda line: line.split(" "))
word_map = words.map(lambda word: (word,1))
word_count = word_map.reduceByKey(lambda x,y: x+y)
word_count.pprint()
```

```python
ssc.start()
ssc.awaitTermination()
```

```python
ssc.stop()
```
